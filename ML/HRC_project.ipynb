<h4>Custnumber Payment Term Number Engineering

```python
data['cust_payment_terms'].unique()
```

```python
temp = data[:20]
```

```python
import re
def extract_alpha_num(df):
    def segg(alpha_num):
        alpha = re.findall(r'\D+', alpha_num)
        num = re.findall(r'\d+', alpha_num)
        alpha = alpha[0]
        alpha = "".join(alpha)
        if len(num)>0:
            num = int(num[0])
        else:
            num = 0
        return pd.Series(
            (alpha,num),
            index = ['payment_term_alpha','payment_term_num']
        )

    sep = df['cust_payment_terms'].apply(segg)
    df[sep.columns] = sep
    
extract_alpha_num(data)
        
```

<h1>Test Train Split

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
test_data=data[data['clear_date_day'].isnull()]
test_data.shape
```

```python
train_data=data[data['clear_date_day'].notnull()]
train_data.shape
```

```python
train_data.drop('due_in_date',axis=1,inplace=True)
```

```python
train_data.info()
```

```python
X = train_data.drop(['clear_date_day','clear_date_month','delay_day','delay_month'],axis=1)
```

```python
print(X.shape)
X.isnull().sum()
```

```python
y = train_data['delay_month']
print(y.shape)
y.isnull().sum()
```

```python
X_train, X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
X_train.isnull().sum()
```

```python
print(X_test.shape)
X_test.isnull().sum()
```

```python
print(y_train.shape)
y_train.isnull().sum()
```

```python
print(y_test.shape)
y_test.isnull().sum()
```

```python
MSE_Score = []
R2_Score = []
Algorithm = []
```

```python
# Fitting Simple Linear Regression to the Training Set
Algorithm.append('Linear Regression')
clf = LinearRegression()
clf.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = clf.predict(X_test)
```

```python
# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted,squared=False))
R2_Score.append(r2_score(y_test, predicted))
```

```python
# Fitting SVR to the Training Set
Algorithm.append('Support Vector Regression')
clf = SVR()
clf.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = clf.predict(X_test)
```

```python
# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted,squared=False))
R2_Score.append(r2_score(y_test, predicted))
```

```python
# Fitting Decision Tree to the Training Set
Algorithm.append('Decision Tree Regressor')
clf = DecisionTreeRegressor()
clf.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = clf.predict(X_test)
```

```python
# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted,squared=False))
R2_Score.append(r2_score(y_test, predicted))
```

```python
# Fitting Random Forest Regressor Tree to the Training Set
Algorithm.append('Random Forest Regressor')
clf = RandomForestRegressor()
clf.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = clf.predict(X_test)
```

```python
# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted,squared=False))
R2_Score.append(r2_score(y_test, predicted))
```

```python
# Fitting XGBoost Regressor to the Training Set
Algorithm.append('XGB Regressor')
clf = xgb.XGBRegressor()
clf.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = clf.predict(X_test)
```

```python
# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted,squared=False))
R2_Score.append(r2_score(y_test, predicted))
```

```python
# Just Combining the Lists into a DataFrame for a Better Visualisation
Comparison = pd.DataFrame(list(zip(Algorithm, MSE_Score, R2_Score)), columns = ['Algorithm', 'RMSE_Score', 'R2_Score'])
```
