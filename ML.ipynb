```python
import pandas as pd
import numpy as np
import matplotlib as mtl
from sklearn.model_selection import train_test_split

from sklearn.feature_selection import VarianceThreshold
```

```python
# load any dataset
data = pd.read_csv('./dataset.csv')
print(data.shape)
data.head()
```

```python
print(data.shape)
for col in data.columns:
    print(col,data[col].nunique())
```

```python
data.drop([col for col in data.columns if data[col].isnull().sum() > 40000], axis=1,inplace=True)
print(data.shape)
data.head()
```

```python
data.drop(['business_code'],axis=1,inplace=True)
print(data.shape)
data.head()
```

```python
data.drop(['name_customer'],axis=1,inplace=True)
print(data.shape)
data.head()
```

```python
data.drop(['buisness_year'],axis=1,inplace=True)
print(data.shape)
data.head()
```

```python
data['invoice_currency'].unique()
```

```python
usdr = 74.70
cadr = 58.81
```

```python
# filter = data["invoice_currency"]=="USD"
data.replace({"invoice_currency": {"USD": usdr, "CAD": cadr}}, inplace=True)
data['new_amount'] = data['total_open_amount']*data['invoice_currency']
data.head()
data['total_open_amount'] = data['new_amount']
data.head()
```

```python
data.drop(['new_amount'],axis=1,inplace=True)
data.drop(['invoice_currency'],axis=1,inplace=True)
```
